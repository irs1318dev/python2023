```python
import pathlib

import numpy as np
import pandas as pd
```

```python
tcpath = pathlib.Path(r"C:\Users\stacy\data\topcoder")
```

```python
train_key = pd.read_csv(tcpath / "trainingAnswers.csv")
train_key
```

```python
train_key[train_key.RunID > 104000].sample(30).sort_values("RunID")
```

```python
runs = pd.concat([
    train_key[train_key.SourceID==source].sample(5) for source in range(7)]
)
```

```python
runs
```

```python
def read_array(run):
    return np.genfromtxt(
        tcpath / "training" / f"{run}.csv", delimiter=",", dtype=np.int32)
```

```python
runs_dict = {
    str(run): read_array(run) for run in runs.RunID.to_list()
}
```

```python
np.savez_compressed("spectra.npz", **runs_dict)
```
